```python
##David Russell 2/5/20
##Course: Data Mining
##Assignment 0
##Summary:  This is a script that will count the number of words in a text file that start with each letter.  The results are 
##           stored in a dictionary, printed to the console, and then displayed in a bar chart using the matplotlib library.
##          

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

#I am making the assumption that the dataset is in the same folder as the jupyter notebook file
dataset_path = 'dataset.txt'

#dictionary to store all letters and their word counts
letters = {
        'a': 0,
        'b': 0,
        'c':0,
        'd':0,
        'e':0,
        'f':0,
        'g':0,
        'h':0,
        'i':0,
        'j':0,
        'k':0,
        'l':0,
        'm':0,
        'n':0,
        'o':0,
        'p':0,
        'q':0,
        'r':0,
        's':0,
        't':0,
        'u':0,
        'v':0,
        'w':0,
        'x':0,
        'y':0,
        'z':0
}
```

```python
#function to create bar chart
def Display_bar_chart(a_dict):
    objects = a_dict.keys()
    bins = np.arange(len(objects))
    count = a_dict.values()

    plt.bar(bins, count, align='center', alpha=0.5)
    plt.xticks(bins, objects)
    plt.ylabel('Word Count')
    plt.title('Words Count by First Letter')

    plt.show()
    
#driver function to count words by first letter in dataset
def Count_first_letters(data_path):
    
    #get one line at a time from the dataset, split by whitespace and store in new_line array
    for line in open('dataset.txt'):
        new_line = line.split()
        
        #iterate through each item in the new_line array and check if its first letter matches 
        #a letter in the letters dict
        for word in new_line:
            for key in letters.keys():
                if word[0].lower() == key:
                    letters[key] += 1
        
            
    #print(letters.items())  for testing

    #format for printing out letter counts
    for key in letters.keys():
        print(key, 'or', key.upper(), ' : ', letters[key], '\n')
    
    #chart function call
    Display_bar_chart(letters)
```

```python

#driver function
Count_first_letters(dataset_path)
```
